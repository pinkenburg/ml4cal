```python
def landau(x, *par):
    w = x - par[0]
    return par[1]*np.exp(-(w+np.exp(-par[4]*w))/par[2])+par[3]
```

```python
import numpy as np
from numpy import loadtxt
map = loadtxt('map.csv', delimiter=',',dtype="i8")
```

```python
def find_index(n):
    for i in range(0,8):
        for j in range(0,8):
            if map[i][j]==n:
                return (i,j)
    return (None, None)
```
