```python
t_offset = 6.17742

def tempfit(x, *par):
    return par[0]*np.interp((x - par[1] + t_offset), template[:,0], template[:,1]) + par[2]
    
```

```python
import numpy as np
from    matplotlib import colors

np.set_printoptions(precision=3, linewidth=80)

with open('./data/8gev_2099_27.npy', 'rb') as f: X = np.load(f)
print(f'''Read an array: {X.shape}''')

cnt_bad         = 0
param_bounds=([20.0, 2.0, 1200.0],[12000.0, 28.0, 2200.0])
indices = range(3, 31, 3)

x       = np.linspace(0, 31, 31, endpoint=False)
x_short = np.arange(3, 31, 3)

# print(np.array(x))
# print(np.array(x_short))


# 3 values: amplitude, time, pedestal

first = True

for i in range(13000): # loop over the data sample
    wave        = X[i][0:31]
    wave_short  = np.take(wave, indices)

    maxindex    =   np.argmax(wave)
    maxval      =   wave[maxindex]

    maxindex_short = np.argmax(wave_short)
    maxval_short   = wave_short[maxindex_short]

    maxindex_short = x[maxindex]
#    print(maxindex, maxval, maxindex_short, maxval_short)

    try:
        popt, _         = scipy.optimize.curve_fit(tempfit, x,          wave,        p0=[float(maxval-1580),         float(maxindex),       1580.0],        bounds = param_bounds)
        popt_short, _   = scipy.optimize.curve_fit(tempfit, x_short,    wave_short,  p0=[float(maxval_short-1580),   float(maxindex_short), 1580.0],        bounds = param_bounds)
    except:
        cnt_bad+=1
        continue

    if first:
        output_array = np.array([np.append(popt,popt_short)])
        first = False
    else:
        output_array = np.append(output_array, [np.append(popt,popt_short)], axis=0)



print(np.average(output_array[:,0]), np.std(output_array[:,0]))

# print(output_array)
#print(output_array[:,1])

plt.rcParams["figure.figsize"] = (30,10)
myTuple = (None, None, None)



vec = (output_array[:,0]-output_array[:,3])/output_array[:,0]
sigma   = ("%.4f" % np.std(vec))
avg     = ("%.4f" % np.average(vec))

fig, myTuple = plt.subplots(1,3)

_ = myTuple[0].hist2d(output_array[:,0], output_array[:,0]-output_array[:,3], bins=(100,100), range=((0.0, 6000.0), (-200.0, 200)), norm=colors.LogNorm(1.0), cmap='plasma')
myTuple[0].set_title("Residuals vs Amplitude", fontweight='bold')
myTuple[0].set_xlabel('Amplitude (ADC counts)')
myTuple[0].set_ylabel('Residuals')
myTuple[0].xaxis.set_zorder(10.0)
myTuple[0].yaxis.set_zorder(10.0)
myTuple[0].grid(True)

_ = myTuple[1].hist((output_array[:,0]-output_array[:,3])/output_array[:,0], bins=100, color="green")
myTuple[1].set_title(f'''Std: {sigma}, Avg: {avg}''', fontweight='bold')
myTuple[1].set_xlabel('Residual/Amplitude')
myTuple[1].set_ylabel('Residual')

myTuple[1].xaxis.set_zorder(10.0)
myTuple[1].yaxis.set_zorder(10.0)
myTuple[1].grid(True)


_ = myTuple[2].hist2d(output_array[:,0], (output_array[:,0]-output_array[:,3])/output_array[:,0], bins=(100,100), range=((0.0, 6000.0), (-0.1, 0.1)), norm=colors.LogNorm(1.0), cmap='plasma')
myTuple[2].set_title("Relative Residuals vs Amplitude", fontweight='bold')
myTuple[2].set_xlabel('Amplitude (ADC counts)')
myTuple[2].set_ylabel('Relative Residuals')
myTuple[2].xaxis.set_zorder(10.0)
myTuple[2].yaxis.set_zorder(10.0)
myTuple[2].grid(True)



#    fit  = tempfit(x, *popt)
```
